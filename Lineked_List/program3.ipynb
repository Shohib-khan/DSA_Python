### WAP TO SOLVE THE GIVEN EXPRESSION INTO POST FIX EVALUATION

```python
def postfix_eval(expression):
    stack =[]
    
    for char in expression:
        if ord(char)>=49 and ord(char)<=57:
            stack.append(int(char))
            
        else:
            n2=stack.pop()
            n1=stack.pop()
            
            if char =='+':
                stack.append((n1+n2))
            elif char =='-':
                stack.append((n1-n2))
            elif char =='*':
                stack.append((n1*n2))
            elif char == '/':
                stack.append(n1/n2)
            elif char =='^':
                stack.append(n1**n2)
    return stack.pop()  


exp ="23*54*+9-"
print(postfix_eval(exp))                             
```

### Harsh is try to arrange him textbooks in a boolshelf that shelf is able to take only the limited number of books than immediately arrange his books in a new shelf automatically
### now consider arrangement of books in a stack if one stack is reached maximum size immediately crete new stack and push the books to the newely created stack


```python
class BOOK_STACK:
    def __init__(self,maxSize):
        self.maxSize =maxSize
        self.s1 =[]
        
    def push(self,val):
        if len(self.s1)>0 and len(self.s1[-1]) <self.maxSize:
            self.s1[-1].append(val)
        else:
            self.s1.append([val])
    def pop(self):
        if len(self.s1<=0):
            print("nothing to pop")
        elif len(self.s1) and len(self.s1[-1])>0:
            self.s1[-1].pop()
        else:
            self.s1.pop()    
                         
    def popAt(self,stk_num):
        if len(self.s1)>0:
            if stk_num <= len(self.s1) and stk_num>0:
                self.s1[stk_num-1].pop()
            else:
                print("No stack is present at given num")
        else:
            print("STACK IS Empty")    
            
    def peek(self):
        if len(self.s1)>0:
            print(self.s1[-1][-1])
        else:
            print("Stack is Empty")
            
    def peekAt(self,stk_num):
        if len(self.s1)>0:
            if stk_num <=len(self.s1) and stk_num>0:
                print(self.s1[stk_num-1][-1])
            else:
                print("No stack num")
        else:
            print("stack has no element")
                                                                
                
                
    def pushAt(self,val,stk_num):
        if stk_num <=len(self.s1) and stk_num>0:
            if len(self.s1[stk_num-1])<self.maxSize:
                self.s1[stk_num-1].append(val)
            else:
                print("stack reached its limit")
        else:
            print("No stack number")                      
                
                                    
    def display(self):
        if len(self.s1) !=0:
            print(self.s1)
        else:
            print("no element")
            
            
            

books =BOOK_STACK(4)  
books.push("Python")
books.push("Java")
books.push("C")
books.push("C++")
books.push("JavaScript")

# Automatically creates second stack
books.push("SQL")
books.push("HTML")
books.push("CSS")
books.push("React")
books.push("Django")

# Automatically creates third stack
books.push("Flask")     
books.display()                          
```
