```python
def sumOfDegit(n):
    if n ==0:
        return 0
    else:
        return n%10+sumOfDegit(n//10)

print(sumOfDegit(81265))    
```

### WAP to convert decimal to binary using recurssion

```python
def dtb(n):
    if n>1:
        dtb(n//2)
    print(n%2,end=" ")
dtb(10)        
```

### Display only even numbers upto 10 using recurssion

```python
def dis_even(n):
    if n<=10:
        print(n,end=" ")
        dis_even(n+2)    
    
dis_even(2)    
```

### Find sum of the given element present in the list using recursive algo

```python
def Sum_Of_Ele(l):
    if len(l)==0:
        return 0
    else:
        return l[0]+Sum_Of_Ele(l[1:])

l=[1,2,3,6,5]
Sum_Of_Ele(l)    
```

### reverse the string using recurssion

```python
def rev_string(s1):
    if len(s1)==0:
        return ""
    else:
        return (s1[-1]+rev_string(s1[:-1]))
s1 ="python"
rev_string(s1)
```

Convert the first and last character of every word to the uppercase using recursion only

```python

def convert_recursive(lst, index=0):
    if index == len(lst):      # Base case
        return
    word = lst[index]
    if len(word) == 1:
        lst[index] = word.upper()
    else:
        lst[index] = word[0].upper() + word[1:-1] + word[-1].upper()
    convert_recursive(lst, index + 1)   # Recursive call
l1 = ["python", "coding", "is", "awesome", "when", "we", "practive"]
convert_recursive(l1)
print(l1)
```
