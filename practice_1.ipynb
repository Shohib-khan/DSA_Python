### WAP to insert single linked list inside a stack data structure

```python
class Node:
    def __init__(self,data):
        self.data =data
        self.addr =None

class Stack_Linked_List:
    def __init__(self):
        self.head =None
        
    def push(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head =newNode
        else:
            newNode.addr = self.head
            self.head =newNode                

    def display(self):
        if self.head is None:
            print("No nodes")
        else:
            temp =self.head
            while temp:
                print(temp.data)    
                temp =temp.addr
            print()
    
    def pop(self):
        if self.head is None:
            print("No nodes in stack")
        else:
            self.head =self.head.addr
    
    def peek_ele(self):
        if self.head is None:
            print("No element to peek")
        else:
            print("Peek node element :",self.head.data)
            

ssl =Stack_Linked_List()
ssl.push('A')
ssl.push('B')                                        
ssl.push('c')                                        
ssl.display()                             
```

### A person arranging his books ,help him to arracnge the books in the form of a stack dara structure 

Arrange the books one above another and if it is reached limited size of 5 then at the book in a 2nd stack automatically ,

so each stack contains 5 books and all the stacks is present in a outer stack , so inner stack as a limited size but outer

stack has no limited size

create a methods like push ,display , pop, and potat



```python
class InnerStack:
    def __init__(self, limit=5):
        self.limit = limit
        self.stack = []

    def push(self, book):
        if len(self.stack) < self.limit:
            self.stack.append(book)
            return True
        return False

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None

    def is_full(self):
        return len(self.stack) == self.limit

    def is_empty(self):
        return len(self.stack) == 0

    def display(self):
        return self.stack


class SetOfStacks:
    def __init__(self, limit=5):
        self.limit = limit
        self.outer_stack = []

    # Push a new book
    def push(self, book):

        # If no stack exists or last stack is full
        if len(self.outer_stack) == 0 or self.outer_stack[-1].is_full():
            self.outer_stack.append(InnerStack(self.limit))

        self.outer_stack[-1].push(book)
        print(f"{book} added successfully.")

    # Pop from last stack
    def pop(self):

        if len(self.outer_stack) == 0:
            print("No books available.")
            return

        book = self.outer_stack[-1].pop()

        # Remove empty stack
        if self.outer_stack[-1].is_empty():
            self.outer_stack.pop()

        print(f"Removed Book: {book}")
        return book

    # Pop from a particular stack
    def popAt(self, index):

        if index < 0 or index >= len(self.outer_stack):
            print("Invalid Stack Number")
            return

        if self.outer_stack[index].is_empty():
            print("Selected stack is empty.")
            return

        book = self.outer_stack[index].pop()

        # Remove empty stack
        if self.outer_stack[index].is_empty():
            self.outer_stack.pop(index)

        print(f"Removed '{book}' from Stack {index + 1}")

    # Display all stacks
    def display(self):

        if len(self.outer_stack) == 0:
            print("No books present.")
            return

        print("\nBooks Arrangement:")

        for i, stack in enumerate(self.outer_stack):
            print(f"Stack {i + 1}: {stack.display()}")


# ---------------- Driver Code ----------------

books = SetOfStacks()

books.push("Python")
books.push("Java")
books.push("C")
books.push("C++")
books.push("JavaScript")

# Automatically creates second stack
books.push("SQL")
books.push("HTML")
books.push("CSS")
books.push("React")
books.push("Django")

# Automatically creates third stack
books.push("Flask")

books.display()

print("\n")

books.pop()

books.display()

print("\n")

books.popAt(0)

books.display()
```
