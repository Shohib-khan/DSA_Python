```python
class QUEUE:
    def __init__(self,maxSize):
        self.maxSize =maxSize
        self.q1 =[]
    
    def IsFull(self):
        if len(self.q1) ==self.maxSize:
            return True
        else:
            return False
    
    def IsEmpty(self):
        if len(self.q1)==0:
            return True
        else:
            return False
    
    def enqueue(self,val):
        if self.IsFull():
            print("queue is full")
        else:
            self.q1.append(val)
    
    def display(self):
        if self.IsEmpty():
            print("no element to display")
        else:
            print(self.q1) 
            
    def dequeue(self):
        if self.IsEmpty():
            print("no element to remove")
        else:
            self.q1.pop(0)
    def peek(self):
        if self.IsEmpty():
            print("No element to peek")
        else:
            print("peek element is ",self.q1[0])                                                   
            
            
q =QUEUE(5)
q.display()   
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)         
q.enqueue(40)         
q.enqueue(50)   
q.display()      
q.dequeue()
q.display()
         
```
